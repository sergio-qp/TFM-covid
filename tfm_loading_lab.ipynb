```python
lab["Days"] = lab["Fecha_Extracci_n_petici_n"]-lab["Fecha__ingreso_"]
lab["Days"]=lab["Days"].dt.days
lab
```

```python
lab["sumHour"] = lab["hora"]+(lab["Days"]*24)
lab
```

```python
descVars = [
    "San-Basófilos",
    "San-Eosinófilos",
    "San-Linfocitos",
    "San-Monocitos",
    "San-Neutrófilos"
]

extendedVars = [
    "Srm-Creatinina",
    "Srm-Urea",
    "Srm-Sodio",
    "Srm-Potasio",
    "Srm-Glucosa",
    # "Srm-Proteina C reactiva (PCR)",
    # "Srm-Procalcitonina (PCT)"
]
#this seems good, only blood is 1465, so thats our best, with the 5 others it drops to like 1407, the last to drop it to 1294 tho, so middle
```

```python
# ctes.loc[ctes["Etiqueta__form_cte_"]==var,"cont"]
clean_lab = []
i=0
for patient in lab["Id_Paciente"].unique():
    temp_ls = dict()
    min_hour = min(lab.loc[lab["Id_Paciente"]==patient]["sumHour"])
    # max_hour = max(lab.loc[lab["Id_Paciente"]==patient]["sumHour"])    
    temp_ls["Id_Paciente"] = patient #to make leftmost col
    temp_ls["sumHour"] = min_hour
    temp_ls=pd.DataFrame([temp_ls])
    # temp_ls["Id_Paciente"] = patient #to fill
    # descVars
    # print(temp_ls)
    for variable in descVars+extendedVars:

        # print(temp_ls)
        # I think what we do is, first, filter to only the first hour, next tag the smaller number in the column as 
        # the absolute value and the higher one as the percentage value and tag them as such


        filter=np.where((lab['Id_Paciente']==patient) & (lab['Desc_Prueba']==variable)) #gross but works great
        thinframe = lab.loc[filter].loc[:,["sumHour","Resultado_n_mero","Resultado_Patol_gico"]]
        
        if len(thinframe["Resultado_n_mero"]) > 0:
            if variable in descVars:
                thinframe[variable+"_absVal_pat"] = thinframe[thinframe["Resultado_n_mero"]==min(thinframe["Resultado_n_mero"])]["Resultado_Patol_gico"]
                thinframe[variable+"_Pct_pat"] = thinframe[thinframe["Resultado_n_mero"]==max(thinframe["Resultado_n_mero"])]["Resultado_Patol_gico"]
                if len([x for x in thinframe[variable + "_absVal_pat"] if pd.notna(x)]) > 0:

                    thinframe[variable+"_absVal"] = min(thinframe["Resultado_n_mero"]) #set smallest value to abs value
                    thinframe[variable+"_Pct"] = max(thinframe["Resultado_n_mero"]) #set largest value to percent
                    thinframe = thinframe.drop(columns=['Resultado_n_mero', 'Resultado_Patol_gico'])

                    thinframe[variable+"_absVal_pat"] = [x for x in thinframe[variable + "_absVal_pat"] if pd.notna(x)][0]
                    thinframe[variable+"_Pct_pat"] = [x for x in thinframe[variable + "_Pct_pat"] if pd.notna(x)][0]
                    thinframe = thinframe.drop_duplicates()
                    # print(thinframe)
                    temp_ls = pd.merge(temp_ls,thinframe,how='left',on=['sumHour'])
            else:
                thinframe[variable+"_val"] = thinframe["Resultado_n_mero"] #try running this when u wake up
                thinframe[variable+"_pat"] = thinframe["Resultado_Patol_gico"]
                thinframe = thinframe.drop(columns=['Resultado_n_mero', 'Resultado_Patol_gico'])
                thinframe = thinframe.drop_duplicates().dropna()


                # print(thinframe)
                # print(temp_ls)
                temp_ls = pd.merge(temp_ls,thinframe,how='left',on=['sumHour'])

    
    clean_lab.append(temp_ls)
    i=i+1
    print(i/len(lab["Id_Paciente"].unique()))
```

So, our suspicion is these different variables could be, one is for absolute count, meaning num * 10^9 per Liter, and the other is percentage of white blood cells, let's see if they add up! Did the math, the larger numbers in all cases add up to exactly 100! awesome!

Also, by looking at it, lets just take the first measurement from each person, so no hourly measurements, just a flat 1 measure for each patient, even if they have more than one, just pick the first one.
